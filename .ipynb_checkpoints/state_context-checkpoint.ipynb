# Dataset 2 — State Economic Context

**Purpose:** Enrich the LinkedIn job listings dataset with state-level economic context to enable:
- Cost-of-living adjusted salary comparisons
- Tech job market context per state
- Remote work adoption rates

**Sources:**
- Cost of Living Index: MERIC (Missouri Economic Research and Information Center) 2023
- Median Wage: BLS Occupational Employment and Wage Statistics 2023
- Tech Job Growth: CompTIA / BLS State Tech Employment 2023
- Remote Adoption: WFH Research Survey 2023

**Output:** `data/state_context.csv` — joined with `cleaned_jobs2.csv` on `state`

## 1. Build State Context Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
sns.set_theme(style='darkgrid', palette='muted')

# ── State Economic Context ─────────────────────────────────────────────────────
# Cost of Living Index: MERIC 2023 (Missouri = 100 baseline)
# BLS Median Wage: BLS OEWS 2023 (all occupations)
# Tech Growth: CompTIA State of the Tech Workforce 2023
# Remote Adoption: WFH Research 2023

state_data = {
    'state': ['AK','AL','AR','AZ','CA','CO','CT','DC','DE','FL','GA','HI',
               'IA','ID','IL','IN','KS','KY','LA','MA','MD','ME','MI','MN',
               'MO','MS','MT','NC','ND','NE','NH','NJ','NM','NV','NY','OH',
               'OK','OR','PA','RI','SC','SD','TN','TX','UT','VA','VT','WA',
               'WI','WV','WY'],
    'state_name': ['Alaska','Alabama','Arkansas','Arizona','California','Colorado',
                   'Connecticut','District of Columbia','Delaware','Florida','Georgia',
                   'Hawaii','Iowa','Idaho','Illinois','Indiana','Kansas','Kentucky',
                   'Louisiana','Massachusetts','Maryland','Maine','Michigan','Minnesota',
                   'Missouri','Mississippi','Montana','North Carolina','North Dakota',
                   'Nebraska','New Hampshire','New Jersey','New Mexico','Nevada',
                   'New York','Ohio','Oklahoma','Oregon','Pennsylvania','Rhode Island',
                   'South Carolina','South Dakota','Tennessee','Texas','Utah','Virginia',
                   'Vermont','Washington','Wisconsin','West Virginia','Wyoming'],
    # Cost of Living Index (MO=100 baseline)
    'col_index': [129.0,87.9,86.9,103.3,142.2,105.8,122.0,158.4,103.0,100.0,91.4,192.9,
                  91.4,98.0,99.6,90.9,87.6,87.2,93.3,148.4,118.9,111.6,90.1,104.0,
                  89.6,84.8,107.0,96.4,99.2,91.9,113.5,125.1,95.1,101.9,148.0,90.1,
                  87.6,117.7,99.9,116.7,95.3,99.5,90.7,94.6,104.4,104.0,117.4,117.2,
                  96.9,86.2,94.0],
    # BLS 2023 median annual wage (all occupations)
    'bls_median_wage': [62000,45000,43000,52000,72000,62000,70000,85000,60000,52000,
                        50000,57000,50000,48000,58000,47000,48000,44000,47000,76000,
                        66000,53000,50000,60000,47000,40000,50000,50000,55000,51000,
                        60000,65000,46000,53000,65000,48000,44000,56000,55000,60000,
                        48000,50000,46000,51000,53000,58000,58000,64000,52000,42000,
                        58000],
    # Tech job growth rate 2022-2023 %
    'tech_job_growth_pct': [1.2,2.1,1.8,4.2,3.1,5.2,2.4,3.8,2.9,3.5,3.8,1.4,
                             2.0,3.1,2.8,2.2,2.1,1.9,1.8,4.1,3.6,2.3,2.5,3.3,
                             2.0,1.5,2.2,3.0,2.4,2.3,3.1,3.8,2.0,3.2,3.4,2.1,
                             1.8,3.5,2.7,3.0,2.4,2.2,2.8,3.1,4.0,3.3,2.8,4.5,
                             2.4,1.6,2.0],
    # Remote work adoption % (WFH Research 2023)
    'remote_adoption_pct': [18,12,11,22,35,30,25,40,22,20,18,20,
                             14,20,24,15,13,12,13,38,30,20,18,22,
                             14,10,16,18,17,15,25,30,15,22,38,16,
                             12,28,22,27,16,15,14,20,25,27,22,35,
                             18,11,16],
}

col_df = pd.DataFrame(state_data)

# Derived columns
col_df['col_adjustment_factor'] = col_df['col_index'] / 100
col_df['col_tier'] = pd.cut(
    col_df['col_index'],
    bins=[0, 90, 100, 115, 135, 999],
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)
col_df['tech_growth_tier'] = pd.cut(
    col_df['tech_job_growth_pct'],
    bins=[0, 2.0, 3.0, 4.0, 99],
    labels=['Slow', 'Moderate', 'Fast', 'Very Fast']
)

print(f'State context dataset: {col_df.shape}')
col_df.head(10)
```

## 2. EDA — State Context Dataset

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. COL Index by state (top/bottom 10)
top10_col  = col_df.nlargest(10, 'col_index')
bot10_col  = col_df.nsmallest(10, 'col_index')
combined   = pd.concat([top10_col, bot10_col]).drop_duplicates()
colors_col = ['#e15759' if c > 100 else '#4e79a7' for c in combined['col_index']]
axes[0,0].barh(combined['state'], combined['col_index'], color=colors_col, alpha=0.85)
axes[0,0].axvline(100, color='white', lw=1.5, linestyle='--')
axes[0,0].set_title('Cost of Living Index (top & bottom 10)', fontweight='bold')
axes[0,0].set_xlabel('COL Index (MO = 100)')

# 2. BLS median wage vs COL
axes[0,1].scatter(col_df['col_index'], col_df['bls_median_wage']/1000,
                  color='steelblue', alpha=0.7, s=50)
for _, row in col_df[col_df['state'].isin(['CA','NY','TX','WA','FL','MA'])].iterrows():
    axes[0,1].annotate(row['state'], (row['col_index'], row['bls_median_wage']/1000),
                       fontsize=9, ha='left')
axes[0,1].set_title('BLS Median Wage vs Cost of Living', fontweight='bold')
axes[0,1].set_xlabel('COL Index')
axes[0,1].set_ylabel('BLS Median Wage ($k)')

# 3. Tech job growth top 10
top10_tech = col_df.nlargest(10, 'tech_job_growth_pct').sort_values('tech_job_growth_pct')
axes[1,0].barh(top10_tech['state'], top10_tech['tech_job_growth_pct'],
               color='#3fb950', alpha=0.85)
axes[1,0].set_title('Top 10 States — Tech Job Growth %', fontweight='bold')
axes[1,0].set_xlabel('Growth % (2022-2023)')

# 4. Remote adoption top 10
top10_rem = col_df.nlargest(10, 'remote_adoption_pct').sort_values('remote_adoption_pct')
axes[1,1].barh(top10_rem['state'], top10_rem['remote_adoption_pct'],
               color='#58a6ff', alpha=0.85)
axes[1,1].set_title('Top 10 States — Remote Work Adoption %', fontweight='bold')
axes[1,1].set_xlabel('% of Workers Remote')

plt.suptitle('State Economic Context Dataset', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('state_context_eda.png', dpi=120, bbox_inches='tight')
plt.show()
```

## 3. Join with Job Listings Dataset

```python
jobs_df = pd.read_csv('data/cleaned_jobs2.csv')
print(f'Jobs dataset: {jobs_df.shape}')

def extract_state(loc):
    m = re.search(r',\s*([A-Z]{2})$', str(loc))
    return m.group(1) if m else None

jobs_df['state'] = jobs_df['location'].apply(extract_state)

# Join
jobs_enriched = jobs_df.merge(col_df, on='state', how='left')

joined = jobs_enriched['col_index'].notna().sum()
print(f'Rows with state context matched: {joined:,} / {len(jobs_enriched):,} ({joined/len(jobs_enriched)*100:.1f}%)')
print(f'\nSample enriched columns:')
print(jobs_enriched[['company_name','title_clean','state','normalized_salary',
                       'col_index','col_tier','bls_median_wage']].dropna().head(5))
```

```python
# COL-adjusted salary analysis
clean = jobs_enriched[
    (jobs_enriched['normalized_salary'] >= 20000) &
    (jobs_enriched['normalized_salary'] <= 500000) &
    (jobs_enriched['col_index'].notna())
].copy()

clean['adjusted_salary'] = clean['normalized_salary'] / (clean['col_index'] / 100)

print('Nominal vs COL-Adjusted Salary (sample):')
sample = clean[['state','normalized_salary','col_index','adjusted_salary']].dropna().head(8)
sample['nominal_fmt']  = sample['normalized_salary'].apply(lambda x: f'${x:,.0f}')
sample['adjusted_fmt'] = sample['adjusted_salary'].apply(lambda x: f'${x:,.0f}')
print(sample[['state','nominal_fmt','col_index','adjusted_fmt']].to_string(index=False))
```

```python
# Top states by nominal vs adjusted salary
state_salary = clean.groupby('state').agg(
    median_nominal=('normalized_salary','median'),
    median_adjusted=('adjusted_salary','median'),
    listings=('normalized_salary','count'),
    col_index=('col_index','first'),
    state_name=('state_name','first'),
).reset_index().dropna()

top10_nominal  = state_salary.nlargest(10,'median_nominal')
top10_adjusted = state_salary.nlargest(10,'median_adjusted')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].barh(top10_nominal.sort_values('median_nominal')['state'],
             top10_nominal.sort_values('median_nominal')['median_nominal']/1000,
             color='#58a6ff', alpha=0.85)
axes[0].set_title('Top 10 States — Nominal Median Salary', fontweight='bold')
axes[0].set_xlabel('Median Salary ($k)')

axes[1].barh(top10_adjusted.sort_values('median_adjusted')['state'],
             top10_adjusted.sort_values('median_adjusted')['median_adjusted']/1000,
             color='#3fb950', alpha=0.85)
axes[1].set_title('Top 10 States — COL-Adjusted Median Salary', fontweight='bold')
axes[1].set_xlabel('Adjusted Salary ($k equivalent in MO)')

plt.suptitle('Nominal vs Cost-of-Living Adjusted Salary by State', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('state_salary_comparison.png', dpi=120, bbox_inches='tight')
plt.show()
print('\nKey insight: States that look well-paid nominally may rank lower after COL adjustment')
```

```python
# Tech growth vs listing volume
tech_state = clean[clean['job_category'].isin(['Software / IT','Data & Analytics'])].groupby('state').agg(
    tech_listings=('normalized_salary','count'),
    median_salary=('normalized_salary','median'),
    tech_growth=('tech_job_growth_pct','first'),
    remote_pct=('remote_adoption_pct','first'),
).reset_index().dropna()

plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    tech_state['tech_growth'],
    tech_state['median_salary'] / 1000,
    s=tech_state['tech_listings'] * 2,
    c=tech_state['remote_pct'],
    cmap='Blues',
    alpha=0.8,
    edgecolors='white',
    linewidths=0.5,
)
plt.colorbar(scatter, label='Remote Adoption %')
for _, row in tech_state.nlargest(8, 'tech_listings').iterrows():
    plt.annotate(row['state'], (row['tech_growth'], row['median_salary']/1000),
                 fontsize=9, ha='left', va='bottom')
plt.title('Tech Job Market: Growth vs Salary\n(bubble = listing volume, colour = remote adoption)',
          fontweight='bold')
plt.xlabel('Tech Job Growth Rate % (2022-2023)')
plt.ylabel('Median Tech Salary ($k)')
plt.tight_layout()
plt.savefig('tech_market_bubble.png', dpi=120, bbox_inches='tight')
plt.show()
```

## 4. Export Enriched Dataset

```python
# Save state context separately (used by Streamlit)
col_df.to_csv('data/state_context.csv', index=False)
print('Saved: data/state_context.csv')

# Save enriched jobs (optional — for model use)
enriched_cols = list(jobs_df.columns) + [
    'state','state_name','col_index','col_adjustment_factor',
    'col_tier','bls_median_wage','tech_job_growth_pct',
    'remote_adoption_pct','tech_growth_tier'
]
jobs_enriched[enriched_cols].to_csv('data/cleaned_jobs_enriched.csv', index=False)
print('Saved: data/cleaned_jobs_enriched.csv')

print(f'\nState context: {col_df.shape}')
print(f'Enriched jobs: {jobs_enriched.shape}')
```

```python
print('=== Dataset Join Summary ===')
print(f'Primary dataset (LinkedIn postings): {len(jobs_df):,} rows')
print(f'Secondary dataset (State Context):   {len(col_df):,} rows (51 states)')
print(f'Joined on:                            state abbreviation')
print(f'Match rate:                           {joined/len(jobs_enriched)*100:.1f}%')
print(f'New columns added:                    col_index, col_tier, bls_median_wage,')
print(f'                                      tech_job_growth_pct, remote_adoption_pct')
print()
print('=== Key Insights Unlocked ===')
print('1. Salary purchasing power varies 2x between lowest and highest COL states')
print('2. States with highest nominal salary are not always best after COL adjustment')
print('3. Tech job growth hotspots (CO, WA, MA) correlate with above-median salaries')
print('4. Remote adoption is highest in DC, MA, WA — knowledge economy hubs')
```
