# Job Listings — Data Cleaning & Feature Engineering
**Source:** `data/postings.csv` — 123,849 raw LinkedIn job listings  
**Output:** `data/cleaned_jobs2.csv` — cleaned, normalised, skill-extracted dataset  

### What this notebook does
1. Data acquisition & inspection
2. Data cleaning (nulls, salary, remote status)
3. **Title normalisation** — collapses 72k → ~57k unique titles by stripping location/salary/work-type noise
4. **NLP skill extraction** — extracts 150+ skills from job description tokens
5. Time feature engineering — extracts listing date for trend analysis
6. Export to cleaned CSV (with skills columns included)

## 4. Description Cleaning & Tokenisation

```python
df['description_clean'] = df['description'].fillna('').str.lower()
df['description_clean'] = df['description_clean'].str.replace('\n', ' ')
df['description_clean'] = df['description_clean'].apply(lambda x: re.sub(r'[^\w\s]', ' ', x))
df['description_clean'] = df['description_clean'].str.replace(r'\s+', ' ', regex=True).str.strip()

# Tokenise
df['tokens'] = df['description_clean'].str.split()

# Remove stopwords
nltk.download('stopwords', quiet=True)
stop_words = set(stopwords.words('english'))

df['tokens'] = df['tokens'].apply(
    lambda words: [w for w in words if w not in stop_words] if isinstance(words, list) else []
)

print('Description cleaning done ✅')
df[['description_clean', 'tokens']].head(3)
```

## 5. Title Cleaning

### Problem
The raw dataset has **72,521 unique job titles** — most are near-duplicates caused by:
- Location suffixes: `Account Executive - Buffalo`
- Salary noise: `Account Executive - $19.2/hr`  
- Parenthetical codes: `Staff Accountant (26391)`
- Work type in title: `Azure Data Engineer (Full Time)`
- ALL CAPS inconsistency: `ASSISTANT STORE MANAGER`
- Abbreviations: `Sr.`, `RN`, `VP`, `HR`

### Solution
Rule-based normalisation using regex — strips noise while preserving the core job title.

```python
def normalise_title(title):
    """
    Normalise a raw job title by:
    1. Lowercasing
    2. Removing parenthetical content (codes, work type, locations)
    3. Removing pipe-separated suffixes
    4. Removing salary mentions
    5. Removing location/work-type/specialism tails after dashes
    6. Deduplicating repeated words (RN - Registered Nurse)
    7. Expanding common abbreviations (Sr → Senior, RN → Registered Nurse)
    8. Stripping numeric codes and stray punctuation
    """
    if not isinstance(title, str):
        return 'unknown'
    
    t = title.strip().lower()

    # 1. Remove pipe and everything after
    t = re.sub(r'\s*\|.*$', '', t)

    # 2. Remove parenthetical content
    t = re.sub(r'\s*\([^)]*\)', '', t)

    # 3. Remove salary patterns: $19/hr, $100k, 19.2/hr
    t = re.sub(r'[\$][\d,.]+[k+]?(\s*(\/hr|\/hour|per hour))?', '', t, flags=re.IGNORECASE)
    t = re.sub(r'\d+[k+]\s*(\/yr|\/year|per year)?', '', t, flags=re.IGNORECASE)

    # 4. Remove % signs and surrounding numeric context
    t = re.sub(r'\d+\s*%\s*\w*', '', t)

    # 5. Remove remote/work-type suffixes after dash
    t = re.sub(
        r'\s*[-–]\s*(remote|on.?site|hybrid|part.?time|full.?time|work from home|'
        r'wfh|contract|temp\w*|seasonal|urgent|now hiring|immediate|new grad|entry.?level).*$',
        '', t, flags=re.IGNORECASE
    )

    # 6. Remove city, ST location tail after dash
    t = re.sub(r'\s*[-–]\s*[a-z][a-z\s]{1,20},\s*[a-z]{2}\b.*$', '', t)

    # 7. Remove trailing dash + up to 3 words (specialism/city descriptors)
    t = re.sub(r'\s*[-–]\s*(\w+[\s-]?){1,3}$', '', t)

    # 8. Remove 'for [long company name/location]'
    t = re.sub(r'\s+for\s+.{8,}$', '', t)

    # 9. Remove 'in [location]' tail
    t = re.sub(r'\s+in\s+[a-z][a-z\s,]+$', '', t)

    # 10. Deduplicate consecutive repeated words
    #     e.g. 'rn registered nurse registered nurse' → 'rn registered nurse'
    words = t.split()
    seen  = []
    for w in words:
        if w not in seen:
            seen.append(w)
    t = ' '.join(seen)

    # 11. Expand abbreviations
    abbrevs = [
        (r'\bsr\.?\b',  'senior'),
        (r'\bjr\.?\b',  'junior'),
        (r'\bmgr\.?\b', 'manager'),
        (r'\bdir\.?\b', 'director'),
        (r'\bvp\b',     'vice president'),
        (r'\bevp\b',    'executive vice president'),
        (r'\bsvp\b',    'senior vice president'),
        (r'\brn\b',     'registered nurse'),
        (r'\blpn\b',    'licensed practical nurse'),
        (r'\bcna\b',    'certified nursing assistant'),
        (r'\bhr\b',     'human resources'),
        (r'\bit\b',     'information technology'),
        (r'\bcfo\b',    'chief financial officer'),
        (r'\bceo\b',    'chief executive officer'),
        (r'\bcto\b',    'chief technology officer'),
        (r'\bcoo\b',    'chief operating officer'),
    ]
    for pattern, replacement in abbrevs:
        t = re.sub(pattern, replacement, t)

    # 12. Remove numeric codes (3+ digit standalone numbers)
    t = re.sub(r'\b\d{3,}\b', '', t)

    # 13. Remove colons
    t = re.sub(r':', ' ', t)

    # 14. Remove leftover special chars (keep / and &)
    t = re.sub(r'[^\w\s/&]', ' ', t)

    # 15. Collapse whitespace
    t = re.sub(r'\s+', ' ', t).strip()

    return t if len(t) > 1 else 'unknown'


# Apply
df['title_clean'] = df['title'].apply(normalise_title)

print('Title normalisation done ✅')
print(f'  Unique titles before: {df["title"].nunique():,}')
print(f'  Unique titles after:  {df["title_clean"].nunique():,}')
reduction = (1 - df['title_clean'].nunique() / df['title'].nunique()) * 100
print(f'  Reduction:            {reduction:.1f}%')
```

## 6. Job Category Classification

```python
def categorize_job(title):
    t = str(title).lower()
    if any(k in t for k in ['engineer', 'developer', 'programmer', 'devops', 'sre', 'software']):
        return 'Software / IT'
    elif any(k in t for k in ['data', 'analyst', 'analytics', 'scientist', 'bi ', 'intelligence']):
        return 'Data'
    elif any(k in t for k in ['sales', 'account executive', 'business development']):
        return 'Sales'
    elif any(k in t for k in ['marketing', 'seo', 'content', 'brand', 'social media']):
        return 'Marketing'
    elif any(k in t for k in ['manager', 'director', 'head of', 'vice president', 'chief', 'lead']):
        return 'Management'
    elif any(k in t for k in ['nurse', 'medical', 'health', 'clinical', 'physician', 'therapist', 'dental']):
        return 'Healthcare'
    elif any(k in t for k in ['finance', 'accountant', 'accounting', 'financial', 'controller', 'tax', 'audit']):
        return 'Finance'
    elif any(k in t for k in ['human resources', 'recruiter', 'talent', 'hr ']):
        return 'HR'
    else:
        return 'Other'

df['job_category'] = df['title_clean'].apply(categorize_job)

print('Job category distribution:')
print(df['job_category'].value_counts())
```

## 7. NLP Skill Extraction

Keyword matching against an expanded skill dictionary (~150 skills across 6 categories).  
Multi-word skills (e.g. `machine learning`, `power bi`) are matched on the joined token text.

```python
SKILL_DICT = {
    'Programming': [
        'python', 'java', 'javascript', 'typescript', 'c++', 'c#', 'go',
        'ruby', 'php', 'swift', 'kotlin', 'scala', 'rust', 'r',
        'perl', 'matlab', 'bash', 'shell', 'vba', 'groovy',
    ],
    'Data & Analytics': [
        'sql', 'excel', 'tableau', 'power bi', 'pandas', 'numpy',
        'spark', 'hadoop', 'dbt', 'looker', 'sas', 'spss', 'alteryx',
        'databricks', 'snowflake', 'redshift', 'bigquery', 'airflow',
        'etl', 'data warehouse', 'data pipeline',
    ],
    'Cloud & DevOps': [
        'aws', 'azure', 'gcp', 'docker', 'kubernetes', 'terraform',
        'jenkins', 'linux', 'ansible', 'git', 'github', 'gitlab',
        'ci/cd', 'devops', 'microservices', 'rest api', 'graphql',
        'nginx', 'redis', 'kafka',
    ],
    'AI / ML': [
        'machine learning', 'deep learning', 'nlp', 'tensorflow',
        'pytorch', 'scikit-learn', 'computer vision', 'llm',
        'neural network', 'reinforcement learning', 'hugging face',
        'langchain', 'openai', 'generative ai', 'prompt engineering',
        'feature engineering', 'model deployment',
    ],
    'Tools & Platforms': [
        'jira', 'confluence', 'salesforce', 'sap', 'quickbooks',
        'servicenow', 'hubspot', 'figma', 'adobe', 'autocad',
        'microsoft office', 'sharepoint', 'powerpoint', 'outlook',
        'slack', 'zoom', 'trello', 'asana', 'monday',
    ],
    'Soft Skills': [
        'communication', 'leadership', 'teamwork', 'collaboration',
        'problem solving', 'critical thinking', 'project management',
        'presentation', 'negotiation', 'mentoring', 'time management',
        'attention to detail', 'customer service', 'stakeholder management',
    ],
}

# Flat list for matching
ALL_SKILLS      = {skill: cat for cat, skills in SKILL_DICT.items() for skill in skills}
SKILL_LIST      = list(ALL_SKILLS.keys())

print(f'Total skills in dictionary: {len(SKILL_LIST)}')
for cat, skills in SKILL_DICT.items():
    print(f'  {cat:<20}: {len(skills)} skills')
```

```python
# Join tokens back to text for multi-word skill matching
df['text_for_skills'] = df['tokens'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')

def extract_skills(text):
    """Match skills from joined token text. Multi-word skills work correctly."""
    found = []
    for skill in SKILL_LIST:
        if skill in text:
            found.append(skill)
    return found

print('Extracting skills (may take ~30 seconds)...')
df['extracted_skills'] = df['text_for_skills'].apply(extract_skills)

# Split into tech vs soft
tech_categories = {'Programming', 'Data & Analytics', 'Cloud & DevOps', 'AI / ML', 'Tools & Platforms'}
df['tech_skills'] = df['extracted_skills'].apply(
    lambda skills: [s for s in skills if ALL_SKILLS.get(s) in tech_categories]
)
df['soft_skills'] = df['extracted_skills'].apply(
    lambda skills: [s for s in skills if ALL_SKILLS.get(s) == 'Soft Skills']
)

coverage = (df['extracted_skills'].map(len) > 0).sum()
print(f'Skill extraction done ✅')
print(f'  Skills found in {coverage:,} / {len(df):,} listings ({coverage/len(df)*100:.1f}%)')
```

```python
# Top 20 skills overall
all_extracted  = df['extracted_skills'].explode().dropna()
skill_counts   = Counter(all_extracted)
top_skills_df  = pd.DataFrame(skill_counts.most_common(20), columns=['Skill', 'Count'])
top_skills_df['Category'] = top_skills_df['Skill'].map(ALL_SKILLS)

cat_colors = {
    'Programming':      '#4e79a7',
    'Data & Analytics': '#59a14f',
    'Cloud & DevOps':   '#f28e2b',
    'AI / ML':          '#b07aa1',
    'Tools & Platforms':'#e15759',
    'Soft Skills':      '#76b7b2',
}
colors = top_skills_df['Category'].map(cat_colors)

plt.figure(figsize=(11, 7))
bars = plt.barh(top_skills_df['Skill'][::-1], top_skills_df['Count'][::-1], color=colors[::-1])
plt.title('Top 20 Most In-Demand Skills', fontweight='bold', fontsize=13)
plt.xlabel('Number of Job Listings')
plt.tight_layout()
plt.show()

top_skills_df
```

```python
# Top skills per job category
df_exploded = df[['job_category', 'extracted_skills']].explode('extracted_skills').dropna()
top_jobs    = df['job_category'].value_counts().head(5).index

fig, axes = plt.subplots(1, len(top_jobs), figsize=(16, 5))

for ax, job in zip(axes, top_jobs):
    skill_series = df_exploded[df_exploded['job_category'] == job]['extracted_skills']
    top = skill_series.value_counts().head(8)
    ax.barh(top.index[::-1], top.values[::-1], color='steelblue', alpha=0.85)
    ax.set_title(job, fontweight='bold', fontsize=10)
    ax.set_xlabel('Mentions')

plt.suptitle('Top Skills per Job Category', fontweight='bold', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Salary premium by skill
clean_sal = df[(df['normalized_salary'] >= 20_000) & (df['normalized_salary'] <= 500_000)]
top_skill_names = [s for s, _ in skill_counts.most_common(15)]

premium_data = []
base_median  = clean_sal['normalized_salary'].median()

for skill in top_skill_names:
    has_skill  = clean_sal[clean_sal['extracted_skills'].apply(lambda s: skill in s)]
    med_salary = has_skill['normalized_salary'].median()
    premium    = med_salary - base_median
    premium_data.append({'skill': skill, 'median_salary': med_salary, 'premium': premium})

premium_df = pd.DataFrame(premium_data).sort_values('premium', ascending=True)

colors_bar = ['#3fb950' if x >= 0 else '#f78166' for x in premium_df['premium']]

plt.figure(figsize=(10, 6))
plt.barh(premium_df['skill'], premium_df['premium'], color=colors_bar, alpha=0.85)
plt.axvline(0, color='white', linewidth=1, linestyle='--')
plt.title('Salary Premium vs Median — by Skill', fontweight='bold')
plt.xlabel('Salary Premium (USD)')
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x/1000:+.0f}k'))
plt.tight_layout()
plt.show()
```

## 8. Export to Cleaned CSV

```python
columns_to_keep = [
    'company_name',
    'title',
    'title_clean',          # ← normalised title (new)
    'job_category',
    'description',
    'description_clean',
    'tokens',
    'extracted_skills',     # ← full skill list (new)
    'tech_skills',          # ← tech skills only (new)
    'soft_skills',          # ← soft skills only (new)
    'location',
    'formatted_experience_level',
    'work_type',
    'remote_status',
    'normalized_salary',
    'listed_date',          # ← listing date (new)
    'listed_month',         # ← month string for grouping (new)
    'listed_week',          # ← week string for fine-grained trends (new)
]

df_clean = df[columns_to_keep].copy()

print(f'Export shape: {df_clean.shape}')
print(f'Columns: {df_clean.columns.tolist()}')

df_clean.to_csv('data/cleaned_jobs2.csv', index=False)
print('\n✅ Saved to data/cleaned_jobs2.csv')
```
